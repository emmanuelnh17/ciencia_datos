## Problema planteado


El Hospital General de Medellín registra múltiples defunciones cada año, pero estos datos no siempre se analizan para comprender las características de la población fallecida, por lo que esto limita la identificación de patrones demográficos y de las principales causas de mortalidad, que es una información clave para la gestión hospitalaria y la salud pública.

Se utilizara la base de datos de “Defunciones ocurridas en el Hospital General de Medellín” ya que es oficial obtenida a partir de los datos públicos del gobierno Colombiano, y a su vez contiene variables demográficas y epidemiológicas suficentes para realizar un analisis exploratorio inicial.

```python
df.columns = df.columns.str.strip().str.lower().str.replace(" ", "_").str.replace("á","a").str.replace("é","e").str.replace("í","i").str.replace("ó","o").str.replace("ú","u").str.replace("ñ","n")

# Variables más importantes
variables_clave = [
    "ano","trimestre","fecha_defuncion","edad_fallecido","sexo_fallecido","estado_conyugal_fallecido","nivel_educativo_fallecido","ocupacion_fallecido","pertenencia_etnica","probable_manera_muerte","causa_directa"
]
print("\nVista rápida de variables clave:")
print(df[variables_clave].head())

# Vista rápida de las variables clave
print("\nEstadísticas descriptivas de variables clave:")
print(df[variables_clave].describe())
```

## Visualizaciones exploratorias básicas

```python
# Histograma: distribución de edad de los fallecidos
plt.figure(figsize=(8,5))
sns.histplot(df['edad_fallecido'].dropna(), bins=15, kde=True, color='salmon')
plt.title("Distribución de la edad de los fallecidos")
plt.xlabel("Edad")
plt.ylabel("Frecuencia")
plt.show()
```

```python
# Gráfico de barras: distribución por sexo de los fallecidos
plt.figure(figsize=(8,5))
sns.countplot(x='sexo_fallecido', data=df, palette='Set2')
plt.title("Distribución de fallecimientos por sexo")
plt.xlabel("Sexo")
plt.ylabel("Número de fallecidos")
plt.show()
```

```python
# Gráfico de barras: probable manera de muerte
plt.figure(figsize=(8,5))
sns.countplot(x='probable_manera_muerte', data=df, palette='pastel')
plt.title("Distribución por probable manera de muerte")
plt.xlabel("Probable manera de muerte")
plt.ylabel("Número de fallecidos")
plt.xticks(rotation=45)
plt.show()
```

```python
# Principales causas directas de defunción
plt.figure(figsize=(10,6))
df['causa_directa'].value_counts().head(10).plot(kind='bar', color='teal')
plt.title("Principales causas directas de defunción")
plt.xlabel("Causa")
plt.ylabel("Número de defunciones")
plt.xticks(rotation=75)
plt.show()


```

```python
# Edad de los fallecidos según sexo
plt.figure(figsize=(8,6))
sns.boxplot(x="sexo_fallecido", y="edad_fallecido", data=df, palette="Set2")
plt.title("Edad de fallecimiento según sexo")
plt.xlabel("Sexo del fallecido")
plt.ylabel("Edad")
plt.show()
```

```python
# Edad de los fallecidos según estado conyugal
plt.figure(figsize=(10,6))
sns.boxplot(x="estado_conyugal_fallecido", y="edad_fallecido", data=df, palette="coolwarm")
plt.title("Edad de fallecimiento según estado conyugal")
plt.xlabel("Estado conyugal")
plt.ylabel("Edad")
plt.xticks(rotation=45)
plt.show()
```

```python
# Edad según la probable manera de muerte
plt.figure(figsize=(10,6))
sns.boxplot(x="probable_manera_muerte", y="edad_fallecido", data=df, palette="pastel")
plt.title("Edad de fallecimiento según probable manera de muerte")
plt.xlabel("Probable manera de muerte")
plt.ylabel("Edad")
plt.xticks(rotation=30)
plt.show()
```

## Interpretaciones Iniciales

###¿Qué tendencias o patrones básicos se observan?

->La media de los fallecidos se concentran en edades de los 60 a los 80 años.

->Las causas de muertes naturales son las que mas abundan, a diferencia de que lo que pensaba antes de analizar estos datos, ya que crei que aparecerian mayormente causas violentas

->Fallecen mas hombres que mujeres

->La principal causa de muerte es un choque septico

->Las mujeres parecen vivir un poco mas de tiempo que los hombres

->El estado conyugal parece influenciar en la edad de fallecimiento


###¿Qué limitaciones tiene la base de datos?

-> Datos sin informacion, indeterminados o en estudio generan ruido a la hora de analizar los datos





###¿Qué hipótesis iniciales podrían plantearse para análisis posteriores?
-> Con solo dos años de datos (2022 y 2023) es difícil hablar de tendencias temporales, por lo que siento que sera mejor analizar cada año por separado y luego hacer comparaciones.

## Conclusiones


->La mayoría de las defunciones ocurren en personas mayores, principalmente entre los 60 y 80 años, con una ligera mayor longevidad en mujeres frente a los hombres.

->Las muertes naturales son predominantes, siendo el choque séptico la principal causa directa, mientras que las muertes violentas son menos frecuentes de lo esperado.

->Los factores demográficos como el sexo y el estado conyugal muestran influencia en los patrones de mortalidad, lo que abre la puerta a análisis más detallados para realizar estudios posteriores durante el desarrollo de este proyecto.
